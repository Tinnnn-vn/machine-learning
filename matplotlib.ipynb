```python
# Create lines in a Plot
import matplotlib.pyplot as plt
import numpy as np
y = np.array([3,8,1,10,5,7])
plt.plot(y, 'o:r')
plt.show()
# Ban cung co the su dung cac duong ke sau: ':', '-', '--', '-.'
# Mau sac: 'r':red, 'b':blue, 'g':green, 'c':cyan, 'm':magenta,... hoac co the su dung gia tri mau thap luc phan
```

```python
# Create label and title
import matplotlib.pyplot as plt
import numpy as np
x = np.array([3,4,5,7,10])
y = np.array([2,5,6,3,9])

plt.title("Data")
plt.xlabel("Value")
plt.ylabel("Speed")

plt.plot(x,y)
plt.show
```

```python
# Add Grid Lines to a Plot
import matplotlib.pyplot as plt
import numpy as np
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Data")
plt.xlabel("Value")
plt.ylabel("Speed")

plt.plot(x,y)
plt.grid()
plt.show()
```

```python
# Display Multiple Plots
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 1)
plt.plot(x,y)
plt.title("Data")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 2)
plt.plot(x,y)
plt.title("Sales")

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 3)
plt.plot(x,y)
plt.title("Income")

plt.suptitle("My Ecomer", color = "red")
plt.show()
```

```python
# Draw scatter plot, scatter()
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.xlabel("Age")
plt.ylabel("Speed")
plt.scatter(x,y)
plt.show
```

```python
# Color each dot
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])
plt.xlabel("Age")
plt.ylabel("Speed")
plt.scatter(x,y, c = colors)
plt.show
```

```python
# Color map
import matplotlib.pyplot as plt
import numpy as np

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,25,25,15])
mylabels = ["Apple","Orange","Banana","Coconut"]
plt.pie(y, labels = mylabels)
plt.show
```
