# Vector Đơn Vị (Unit vector)
Trong đại số tuyến tính, vector đơn vị là một vector có độ dài (hoặc chuẩn - norm) đúng bằng 1.

Mục đích duy nhất của một vector đơn vị là chỉ ra phương và hướng trong không gian mà không làm ảnh hưởng đến độ lớn của các phép tính toán khác (vì độ lớn của nó bằng 1, khi nhân với bất kỳ số nào cũng giữ nguyên giá trị của số đó).

Công thức:
$$\hat{v} = \frac{\vec{v}}{\|\vec{v}\|}$$

```python
# Vector đơn vị (Unit vector)
def normalize(v1):
    norm_v1 = np.linalg.norm(v1)
    if norm_v1 == 0:
        return v1
    return v1/norm_v1
v1 = ([2,3,4])
v_unit = normalize(v1)
print(f"Unit vector v1: {v_unit}")
print(f"Norm of Unit vector v1: {np.linalg.norm(v_unit)}")
```

# Tích Vô Hướng (Dot Product)
Dot product đo mức độ hai vector "cùng hướng" nhau.
- Nếu dot product lớn dương: hai vector khá cùng hướng.
- Nếu gần 0: hai vector gần vuông góc.
- Nếu âm: hai vector có xu hướng ngược hướng.
- Công thức:
$$\vec{a} \cdot \vec{b} = \|\vec{a}\| \cdot \|\vec{b}\| \cdot \cos(\theta)$$

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
dot_v = np.dot(v1,v2)
print(f"Dot product of vector: {dot_v}")
```

# Tích Có Hướng (Cross Product)
Tích có hướng là phép toán chỉ định nghĩa trong không gian 3 chiều (3D). Kết quả của phép toán này là một vector mới hoàn toàn.
- Công thức:
$$\|\vec{c}\| = \|\vec{a}\| \cdot \|\vec{b}\| \cdot \sin(\theta)$$

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
cross_v = np.cross(v1,v2)
print(f"Cross product of vector: {cross_v}")
```

# Góc Giữa 2 Vector
Góc giữa hai vector (được ký hiệu là $\theta$ hoặc $(\vec{a}, \vec{b})$) là góc được tạo ra khi ta đưa hai vector đó về cùng một gốc. Góc này giúp chúng ta xác định mối quan hệ về mặt hình học, xu hướng và hướng đi tương đối giữa hai vector trong không gian.

Góc giữa hai vector luôn được giới hạn trong khoảng từ $0^\circ$ đến $180^\circ$ (hoặc từ $0$ đến $\pi$ radian)
- Từ công thức hình học của tích vô hướng:
$$\vec{a} \cdot \vec{b} = \|\vec{a}\| \cdot \|\vec{b}\| \cdot \cos(\theta)$$
- Ta có thể suy ra công thức tính $\cos(\theta)$:
$$\cos(\theta) = \frac{\vec{a} \cdot \vec{b}}{\|\vec{a}\| \cdot \|\vec{b}\|}$$

```python
v1 = np.array([1, 0])
v2 = np.array([0, 1])

# Công thức:
# cos(theta) = dot(v1, v2) / (norm(v1) * norm(v2))
cos_theta = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

# arccos trả về radian, ta đổi sang độ bằng np.degrees
theta_rad = np.arccos(cos_theta)
theta_deg = np.degrees(theta_rad)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"cos(theta) = {cos_theta}")
print(f"Góc giữa v1 và v2: {theta_deg}"" độ")
```

# BROADCASTING - CƠ CHẾ RẤT QUAN TRỌNG CỦA NUMPY
Broadcasting cho phép NumPy tự mở rộng shape nhỏ hơn để tính toán với shape lớn hơn.

```python
X = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

bias = np.array([10, 20, 30])

print("X:")
print(X)
print("bias:")
print(bias)

print("X + bias:")
print(X + bias)
```

Trong neural network:
- output = input @ weight + bias

# Ma Trận Vuông - Định Thức - Rank - Ma Trận Nghịch Đảo

```python
# Tạo ma trận vuông 2x2
a = np.array([[1, 2],
              [3, 4]])

print("Ma trận a:")
print(a)

# 1. Định thức của ma trận
det_a = np.linalg.det(a)

print("Định thức của a:")
print(det_a)

# 2. Hạng của ma trận
rank_a = np.linalg.matrix_rank(a)

print("Hạng của ma trận a:")
print(rank_a)

# 3. Ma trận nghịch đảo
inv_a = np.linalg.inv(a)

print("Ma trận nghịch đảo của a:")
print(inv_a)
```

# Giải Hệ Phương Trình Tuyến Tính
Ví dụ:
2x + y = 8,
x + 3y = 13

Viết dạng ma trận:
A @ solution = b

```python
A = np.array([
    [2, 1],
    [1, 3]
])

b = np.array([8, 13])

solution = np.linalg.solve(A, b)

print("A:")
print(A)
print(f"b: {b}")
print(f"Nghiệm [x, y]: {solution}")

# Kiểm tra lại:
print(f"A @ solution = {A @ solution}")
```

Lưu ý:
Nên dùng np.linalg.solve(A, b)
thay vì tự tính np.linalg.inv(A) @ b.

Vì solve thường ổn định và hiệu quả hơn.

# Giá Trị Riêng & Vector Riêng (EIGENVALUES VÀ EIGENVECTORS)
Ý nghĩa đơn giản: 
Một eigenvector là vector mà khi bị ma trận A biến đổi, nó không bị xoay hướng, chỉ bị kéo dài hoặc thu ngắn.

A @ v = lambda * v

lambda là eigenvalue.

v là eigenvector.

```python
A = np.array([
    [2, 0],
    [0, 3]
])

eigenvalues, eigenvectors = np.linalg.eig(A)

print("A:")
print(A)
print("Eigenvalues:")
print(eigenvalues)
print("Eigenvectors:")
print(eigenvectors)
```

# Hồi Quy Tuyến Tính Với Numpy

```python
# Ta có dữ liệu:
# Diện tích nhà -> giá nhà
# Đây chỉ là dữ liệu ví dụ nhỏ để học
X = np.array([
    [30],
    [40],
    [50],
    [60],
    [70]
], dtype=float)

y = np.array([300, 400, 500, 600, 700], dtype=float)

# Mô hình tuyến tính:
# y_pred = X @ w + b
# Để đưa bias b vào phép nhân ma trận, ta thêm một cột toàn số 1 vào X.

ones = np.ones((X.shape[0], 1))
X_bias = np.hstack([X, ones])

print("X_bias:")
print(X_bias)

# Công thức nghiệm bình phương tối thiểu:
# theta = (X^T X)^(-1) X^T y
# theta gồm [w, b]
theta = np.linalg.inv(X_bias.T @ X_bias) @ X_bias.T @ y

w = theta[0]
b = theta[1]

print(f"w = {w}")
print(f"b = {b}")

# Dự đoán giá nhà có diện tích 80
x_new = np.array([[80, 1]])
prediction = x_new @ theta

print(f"Dự đoán giá nhà diện tích 80: {prediction[0]}")
```

Lưu ý:
- Với bài toán thật, ta thường dùng np.linalg.lstsq hoặc thư viện ML.
- Công thức inverse ở trên dùng để học bản chất toán học.
